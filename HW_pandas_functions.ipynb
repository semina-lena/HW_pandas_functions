### Задание 1
Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
df_ratings = pd.read_csv('ratings.csv')
```

```python
def movie_class(rating):
    if rating <= 2:
        value = 'низкий рейтинг'
    if rating >2 and rating <= 4:
        value = 'средний рейтинг'
    if rating >4 and rating <= 5:
        value = 'высокий рейтинг'
    return value
```

```python
df_ratings['class'] = df_ratings['rating'].apply(movie_class)
df_ratings.head(30)
```

```python
def region(row):
    for reg, cities in geo_data.items():
        for city in cities:
            if city in row['keyword']:
                reg_value =  reg
    try:
        res = reg_value        
    except:
        res = 'undefined'
    return res
```

```python
df_keywords['region'] = df_keywords.apply(region, axis=1)
df_keywords.region.value_counts()
```
